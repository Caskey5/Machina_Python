# Vježba OOP - bolnica

```python
# Print - Dobro došli u bolnicu
# Input - Šta trebate? primjer unosa -> operacija
                                #    -> pregled

# Ako pacjent treba operaciju - Odjel_operacija
# Input - Potrebna operacija , primjer unosa -> kralježnica
                                #          -> noga

# Operacija kralježnice na odjel neurokirurgije (Operacija 1)
# Operacija noge na odjel ortopedije (Operacija 2)

# Ako pacijent treba pregled - Odjel_pregled
# Input - Potreban pregled , primjer unosa -> lječnićki pregled
                                #          -> psihološki pregled
# Lječnički pregled (Pregled 1)
# Psihološki pregled (Pregled 2)

# Doktori za određeni zahvat:
# Operacija 1 -> Neurokirurg -> Podaci: Ime, prezime, završena škola
# Operacija 2 -> Ortoped -> Podaci: Ime, prezime, završena škola
# Pregled 1 ->  Lječnik -> Podaci: Ime, prezime, završena škola
# Pregled 2 -> Psihijatar -> Podaci: Ime, prezime, završena škola

# Potrebni podaci pacijenta: ime, prezime, oib, povjest_bolesti, broj_zdravstvene_iskaznice

# Moja zamisao coda:
# Staviti ime, prezime, oib pacijenta u listu.
# Usmijeriti pacijenta na potreban odijel
# Dodati pacijenta u listu -> Prezime, ime, oib, zahvat koji je vršen na pacijentu.
# Odrediti doktora za potreban zahvat
# Otpustiti pacijenta -> Maknuti pacijenta iz liste
# Output bi trebao izgledati ovako:
# - ime i prezime doktora
# - zahvat koji je rađen na (ime i prezime pacijenta)

# 1. Napravi sve potrebne klase 
# 2. Napravi sve potrebne definicije

from __future__ import annotations

class Osoba: 
    def __init__(
        self,
        ime: str,
        prezime: str,
        oib: int
    ):
        self.ime = ime
        self.prezime = prezime
        self.oib = self.provjeri_oib(oib)

    def provjeri_oib(self, oib):
        if len(str(oib)) == 11:
            return oib
        else:
            print("Nevaljan OIB")
            return None
        
class Pacjenti(Osoba):
    id_brojač = 1  # ID brojač

    def __init__(
        self, 
        ime: str, 
        prezime: str, 
        oib: int, 
        zdravstvena_iskaznica: int, 
        zahvat: str
    ):
        super().__init__(ime, prezime, oib)
        self.zdravstvena_iskaznica = self.provijeri_zdravstvenu(zdravstvena_iskaznica)
        self.zahvat = zahvat
        self.id = Pacjenti.id_brojač  # Svakom pacijentu dajem jedinstven ID
        Pacjenti.id_brojač += 1  # Povećavam brojač

    def provijeri_zdravstvenu(self, zdravstvena_iskaznica):
        if len(str(zdravstvena_iskaznica)) == 9:
            return zdravstvena_iskaznica
        else:
            print("Nevažeči broj zdravstvene iskaznice")
            return None
        
    def operacije(operacija: str) -> str:
        operacija_dict: dict[str, str] = {
            "kralježnica": "Operacija kralježnice na odjelu neurokirurgije",
            "noga": "Operacija noge na odjelu ortopedije"
    }
        return operacija_dict.get(operacija.lower(), "Nepoznata operacija, provjerite unos.")

    def pregledi(pregled: str) -> str:
        pregled_dict: dict[str, str] = {
            "lječnički pregled": "Lječnički pregled",
            "psihološki pregled": "Psihološki pregled"
    }
        return pregled_dict.get(pregled.lower(), "Nepoznat pregled, provjerite unos.")
    
    def unos_pacijenta(): 
        ime = input("Unesite ime pacijenta: ")
        prezime = input("Unesite prezime pacijenta: ")

        oib = input("Unesite OIB pacijenta (11 znamenki): ").strip()
        while len(oib) != 11 or not oib.isdigit():
            print("OIB mora imati 11 znamenki i biti broj. Pokušajte ponovo.")
            oib = input("Unesite OIB pacijenta (11 znamenki): ").strip()
        oib = int(oib)

        zdravstvena_iskaznica = input("Unesite broj zdravstvene iskaznice pacijenta (9 znamenki): ").strip()
        while len(zdravstvena_iskaznica) != 9 or not zdravstvena_iskaznica.isdigit():
            print("Broj zdravstvene iskaznice mora imati 9 znamenki i biti broj. Pokušajte ponovo.")
            zdravstvena_iskaznica = input("Unesite broj zdravstvene iskaznice pacijenta (9 znamenki): ").strip()
        zdravstvena_iskaznica = int(zdravstvena_iskaznica)

        zahvat = input("Unesite potreban zahvat (operacija ili pregled): ")

        pacijent = Pacjenti(ime, prezime, oib, zdravstvena_iskaznica, zahvat)
        return pacijent
    
    def otpust_pacijenta(pacijenti: list[Pacjenti], pacijent_id: int) -> list:
        pacijent = None
        for p in pacijenti:
            if p.id == pacijent_id:
                pacijent = p
                break
        if pacijent:
            pacijenti.remove(pacijent)
            print(f"\nOtpust: {pacijent.ime} {pacijent.prezime} (OIB: {pacijent.oib})")
        else:
            print("Ne postoji pacijent s tim ID-em.")
        return pacijenti

    def __repr__(self):
        return f"{self.id}: {self.prezime} {self.ime}: OIB: ({self.oib})\nBroj zdravstvene iskaznice: ({self.zdravstvena_iskaznica})\nPotreban zahvat: {self.zahvat}"


class Neurokirurg(Osoba):
    def __init__(
        self,
        ime: str,
        prezime: str,
        oib: int,
        završena_škola: str,
        zahvat: str
    ):
        super().__init__(ime, prezime, oib)
        self.završena_škola = završena_škola
        self.zahvat = zahvat

    def __repr__(self):
        return f"Neurokirurg: {self.ime} {self.prezime}\nZahvat: {self.zahvat}  "


class Ortoped(Osoba):
    def __init__(
        self,
        ime: str,
        prezime: str,
        oib: str,
        završena_škola: str,
        zahvat: str
    ):
        super().__init__(ime, prezime, oib)
        self.završena_škola = završena_škola
        self.zahvat = zahvat

    def __repr__(self):
        return f"Ortoped: {self.ime} {self.prezime}\nZahvat: {self.zahvat}  "


class Lječnik(Osoba):
    def __init__(
        self,
        ime: str,
        prezime: str,
        oib: int,
        završena_škola: str,
        zahvat: str
    ):
        super().__init__(ime, prezime, oib)
        self.završena_škola = završena_škola
        self.zahvat = zahvat

    def __repr__(self):
        return f"Lječnik: {self.ime} {self.prezime}\nZahvat: {self.zahvat}  "


class Psihijatar(Osoba):
    def __init__(
        self,
        ime: str,
        prezime: str,
        oib: int,
        završena_škola: str,
        zahvat: str
    ):
        super().__init__(ime, prezime, oib)
        self.završena_škola = završena_škola
        self.zahvat = zahvat

    def __repr__(self):
        return f"Psihijatar: {self.ime} {self.prezime}\nZahvat: {self.zahvat}  "

N = Neurokirurg("Ivan", "Nikolić", 10987654321, "Medicinski fakultet", "kralježnica")
LJ = Lječnik("Luka", "Filipović", 25981367895, "Medicinski fakultet", "lječnički pregled")
O = Ortoped("Antonio", "Halar", 55988769693, "Medicinski fakultet", "noga")
P = Psihijatar("Ivano", "Cindrić", 10894555572, "Medicinski fakultet", "psihološki pregled")

# TESTIRANJE KODA

def main():
    print("Dobro došli u bolnicu!")

    # Lista pacijenata
    pacijenti: list[Pacjenti] = [
        Pacjenti("Ivana", "Mileusnić", 12345678910, 123456789, "lječnički pregled"),
        Pacjenti("Nikola", "Duvnjak", 18375671910, 123456789, "noga")
    ]

    # Biranje pacijenta
    while True:
        print("\nDostupni pacijenti: ")
        for pacijent in pacijenti:
            print(f"{pacijent.id}. {pacijent.ime} {pacijent.prezime} (OIB: {pacijent.oib})")

        pacijent_id = int(input("\nUnesite ID pacijenta kojeg želite odabrati (ili 0 za dodavanje novog): ").strip())
        if pacijent_id == 0:
            pacijent = Pacjenti.unos_pacijenta()
            pacijenti.append(pacijent)  # Dodaj novog pacijenta
            print(f"\nNovi pacijent {pacijent.ime} {pacijent.prezime} je dodan.")
        elif any(p.id == pacijent_id for p in pacijenti):
            pacijent = next(p for p in pacijenti if p.id == pacijent_id)
            break
        else:
            print("Neispravan ID, pokušajte ponovo.")

    # Odabir usluge
    izbor = input(f"\nŠto treba {pacijent.ime}? (operacija/pregled): ").strip().lower()

    if izbor == "operacija":
        operacija = input("Koju operaciju treba? (kralježnica/noga): ").strip().lower()
        if operacija == "kralježnica":
            doktor = N  # Neurokirurg izvodi operaciju kralježnice
        elif operacija == "noga":
            doktor = O  # Ortoped izvodi operaciju noge
        else:
            print("Nepoznata operacija, pokušajte ponovo.")
            return

        print(f"\n{doktor.ime} {doktor.prezime} je izveo operaciju na {operacija}.")
        print(f"Pacijent: {pacijent.ime} {pacijent.prezime} (OIB: {pacijent.oib})")

    elif izbor == "pregled":
        pregled = input("Koji pregled treba? (lječnički pregled/psihološki pregled): ").strip().lower()
        if pregled == "lječnički pregled":
            doktor = LJ  # Lječnik izvodi lječnički pregled
        elif pregled == "psihološki pregled":
            doktor = P  # Psihijatar izvodi psihološki pregled
        else:
            print("Nepoznat pregled, pokušajte ponovo.")
            return

        print(f"\n{doktor.ime} {doktor.prezime} je obavio {pregled}.")
        print(f"Pacijent: {pacijent.ime} {pacijent.prezime} (OIB: {pacijent.oib})")

    # Nakon operacije, otpust pacijenta
    pacijenti = Pacjenti.otpust_pacijenta(pacijenti, pacijent.id)

    # Ispis dostupnih pacijenata nakon otpusta
    print("\nDostupni pacijenti nakon otpusta:")
    for pacijent in pacijenti:
        print(f"{pacijent.id}. {pacijent.ime} {pacijent.prezime} (OIB: {pacijent.oib})")


main()


```
